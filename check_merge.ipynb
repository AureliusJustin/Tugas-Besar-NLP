```python
# Merge the three datasets
df_0_1999 = pd.read_csv("dataset_translated_0_1999.csv")
df_2000_3999 = pd.read_csv("dataset_translated_2000_3999.csv")
df_4000_5999 = pd.read_csv("dataset_translated_4000_5999.csv")

# Combine the english_translation columns
# Start with the base dataset (0-1999)
merged_df = df_0_1999.copy()

# Update rows 2000-2999 with translations from second dataset
merged_df.loc[2000:3999, 'english_translation'] = df_2000_3999.loc[2000:3999, 'english_translation'].values

# Update rows 3000-3999 with translations from third dataset
merged_df.loc[4000:5999, 'english_translation'] = df_4000_5999.loc[4000:5999, 'english_translation'].values
# Save the merged dataset
merged_df.to_csv("dataset_translated_0_5999.csv", index=False)

print(f"Rows with translations: {merged_df['english_translation'].notna().sum()}")
print(f"\nSample check:")
print(f"Row 0: {merged_df.iloc[0]['english_translation']}")
print(f"Row 1999: {merged_df.iloc[1999]['english_translation']}")
print(f"Row 2000: {merged_df.iloc[2000]['english_translation']}")
print(f"Row 3999: {merged_df.iloc[3999]['english_translation']}")
print(f"Row 4000: {merged_df.iloc[4000]['english_translation']}")
print(f"Row 5999: {merged_df.iloc[5999]['english_translation']}")
```
