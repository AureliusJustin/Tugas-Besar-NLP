```python
# Check remaining chunks
chunks = [
    ("dataset_translated_2000_2999.csv", "2000-2999"),
    ("dataset_translated_3000_3999.csv", "3000-3999"), 
    ("dataset_translated_4000_4999.csv", "4000-4999"),
    ("dataset_translated_5000_5999.csv", "5000-5999")
]

for filename, range_desc in chunks:
    df = pd.read_csv(filename)
    print(f"Chunk {range_desc}:")
    print(f"Shape: {df.shape}")
    print(f"First row: {df.iloc[0][['id', 'english_translation']]}")
    print(f"Last row: {df.iloc[-1][['id', 'english_translation']]}")
    print()
```

```python
# Properly merge chunks where each file contains full dataset but only specific rows have translations
chunk_files = [
    ("dataset_translated_0_999.csv", 0, 999),
    ("dataset_translated_1000_1999.csv", 1000, 1999), 
    ("dataset_translated_2000_2999.csv", 2000, 2999),
    ("dataset_translated_3000_3999.csv", 3000, 3999),
    ("dataset_translated_4000_4999.csv", 4000, 4999),
    ("dataset_translated_5000_5999.csv", 5000, 5999)
]

# Start with the first chunk as base
base_df = pd.read_csv(chunk_files[0][0])
print(f"Base dataset shape: {base_df.shape}")

# For each subsequent chunk, copy only the translated rows for that range
for filename, start_idx, end_idx in chunk_files[1:]:
    chunk_df = pd.read_csv(filename)
    
    # Copy translations for the specific range from this chunk
    translations_to_copy = chunk_df.loc[start_idx:end_idx, 'english_translation']
    valid_translations = translations_to_copy.notna()
    
    print(f"From {filename}: copying {valid_translations.sum()} translations for rows {start_idx}-{end_idx}")
    
    # Update the base dataset with these translations
    base_df.loc[start_idx:end_idx, 'english_translation'] = translations_to_copy

merged_df = base_df
print(f"\nFinal merged dataset shape: {merged_df.shape}")

# Save the merged dataset
merged_df.to_csv("dataset_cleaned_final_translated.csv", index=False)
print("Saved merged dataset as 'dataset_cleaned_final_translated.csv'")

# Verify the merge
print(f"\nVerification:")
print(f"Total rows: {len(merged_df)}")
print(f"Rows with translations: {merged_df['english_translation'].notna().sum()}")
print(f"Rows without translations: {merged_df['english_translation'].isna().sum()}")

# Sample check at chunk boundaries
print(f"\nSample check at boundaries:")
check_indices = [0, 999, 1000, 1999, 2000, 2999, 3000, 3999, 4000, 4999, 5000, 5999]
for i in check_indices:
    if i < len(merged_df):
        translation = merged_df.iloc[i]['english_translation']
        has_translation = "✓" if pd.notna(translation) else "✗"
        preview = str(translation)[:50] + "..." if pd.notna(translation) and len(str(translation)) > 50 else str(translation)
        print(f"Row {i}: {has_translation} {preview}")
```
