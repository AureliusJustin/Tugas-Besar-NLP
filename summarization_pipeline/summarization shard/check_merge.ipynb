```python
import pandas as pd
```

```python
df_0_999 = pd.read_csv("dataset_summaries_0_999.csv")
print(df_0_999.iloc[0][["id", "summary"]])
print(df_0_999.iloc[999][["id", "summary"]])
```

```python
# Merge the three datasets
df_0_999 = pd.read_csv("dataset_summaries_0_999.csv")
df_1000_1999 = pd.read_csv("dataset_summaries_1000_1999.csv")
df_2000_2999 = pd.read_csv("dataset_summaries_2000_2999.csv")
df_3000_3999 = pd.read_csv("dataset_summaries_3000_3999.csv")
df_4000_4999 = pd.read_csv("dataset_summaries_4000_4999.csv")
df_5000_5999 = pd.read_csv("dataset_summaries_5000_5999.csv")

# Combine the summaries columns
# Start with the base dataset (0-999)
merged_df = df_0_999.copy()
merged_df.loc[1000:1999, 'summary'] = df_1000_1999.loc[1000:1999, 'summary'].values
merged_df.loc[2000:2999, 'summary'] = df_2000_2999.loc[2000:2999, 'summary'].values
merged_df.loc[3000:3999, 'summary'] = df_3000_3999.loc[3000:3999, 'summary'].values
merged_df.loc[4000:4999, 'summary'] = df_4000_4999.loc[4000:4999, 'summary'].values
merged_df.loc[5000:5999, 'summary'] = df_5000_5999.loc[5000:5999, 'summary'].values

# Save the merged dataset
merged_df.to_csv("dataset_summaries.csv", index=False)

print(f"Rows with summaries: {merged_df['summary'].notna().sum()}")
```
